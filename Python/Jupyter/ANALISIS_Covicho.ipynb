```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ajustes de la configuracion en pandas para el formato cientifico a uno amigable
pd.options.display.float_format = "{:,.4f}".format
```

```python
pf = pd.read_csv("./DataSets/covid_19_clean_complete.csv")
pf.sample(20)
```

```python
pf["Date"] = pd.to_datetime(pf["Date"])
pf.sample(20)
```

```python
list(pf)
pf = pf[['Country/Region',
 'Date',
 'Confirmed',
 'Deaths',
 'Recovered',
 'Active']]

pf
```

```python
pfTime = pf.groupby("Date").sum()
pfTime.head()
```

```python
sample1 =  pfTime["Confirmed"]
sample2 =  pfTime["Deaths"]
```

```python
sample1 - sample2
```

Diferencias de la fila anterior con la actual

```python
dif = pfTime.diff().fillna(0)
dif
```

```python
dif.cumsum()
```

```python
dif.resample('D').sum(min_count=1)
```

```python
df_cum = pfTime.resample('12h').sum(min_count=1)
```

```python
df_cum = df_cum.interpolate()
df_cum
```

```python
df_cum['Rate'] = 1 - df_cum['Deaths'] / df_cum['Confirmed']
```

```python
df_cum = df_cum.reset_index()
```

```python
#Tasa de supervivencia
sr = df_cum.groupby(pd.Grouper(key='Date', freq='1D'))[['Rate']].mean()
```

```python
sr.plot()
#sr.rolling(window=7).mean().plot()
```

```python
#Promedios con ventanas de frecuencias
sr.rolling(window=1).mean().plot(title='Ratio de supervivencia')
```

```python
sr.rolling(window=1).std().plot()
```

```python
pf_time = pf.groupby(['Country/Region', pd.Grouper(key='Date', freq='1D')]).sum()
pf_time
```

```python
pf_mexico = pf_time.loc['Mexico', :]
pf_mexico
```

```python
import matplotlib.pyplot as plt
```

```python
pf_mexico.plot( figsize = (10, 7), title='COVID-19')
plt.xlabel('Date')
plt.ylabel('People')
```

```python
pf_mexico.plot( figsize = (10, 7), title='COVID-19', legend=False, style=['r-', 'g--', 'b:*']).legend(['1', 'Dos', 'Tres'])
plt.xlabel('Date')
plt.ylabel('People')
```

```python
df_mx_mont = pf_mexico.resample('M').max()
df_mx_mont.plot(figsize=[10, 7],kind='barh', stacked=True)
```

```python
df_mx_mont['Tratamiento'] = df_mx_mont['Confirmed'] - df_mx_mont['Deaths'] - df_mx_mont['Recovered']
df_mx_mont
```

```python
#Sacando los 0 de las columnas
df_mx_mont = df_mx_mont[(df_mx_mont.T != 0).any()]
df_mx_mont
```

```python
df_mx_mont[['Confirmed','Deaths', 'Recovered', 'Tratamiento']].T.plot(figsize=(10, 7), kind='pie', subplots=True)
```
